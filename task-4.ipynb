```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
start = np.random.random() * np.random.choice([-100, 100])
finish = np.random.random() * 100 + abs(start)
```

```python
k = int(input('Введите значение k '))
x = np.linspace(start, finish, 100)
plt.plot(x, np.cos(x))
plt.plot(x, np.cos(k*x));
```
